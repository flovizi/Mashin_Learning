```python
# выведем выбросы data['game_id'] с помощью фильтра Хэмпеля
# опишем функцию, которая заменяет на Nan значения больше 3-х медианных абсолютных отклонения

def hampel(vals_orig):
    vals = vals_orig.copy()    
    difference = np.abs(vals.median()-vals)
    median_abs_deviation = difference.median()
    threshold = 3 * median_abs_deviation
    outlier_idx = difference > threshold
    vals[outlier_idx] = np.nan
    return(vals)

hampel(data['game_id'])
```

**Тест Граббса**

Автор реализации теста Граббса и теста Рознера для Python
Опишем три функции: первая находит значение критерия Граббса и максимальное значение в наборе данных, вторая — критическое значение с учётом объёма выборки и уровня значимости, а третья проверяет, является ли значение с максимальным индексом выбросом

```python
from scipy import stats

def grubbs_stat(y):
    std_dev = np.std(y)
    avg_y = np.mean(y)
    abs_val_minus_avg = abs(y - avg_y)
    max_of_deviations = max(abs_val_minus_avg)
    max_ind = np.argmax(abs_val_minus_avg)
    Gcal = max_of_deviations / std_dev
    print(f"Grubbs Statistics Value: {Gcal}")
    return Gcal, max_ind

def calculate_critical_value(size, alpha):
    t_dist = stats.t.ppf(1 - alpha / (2 * size), size - 2)
    numerator = (size - 1) * np.sqrt(np.square(t_dist))
    denominator = np.sqrt(size) * np.sqrt(size - 2 + np.square(t_dist))
    critical_value = numerator / denominator
    print(f"Grubbs Critical Value: {critical_value}")
    return critical_value

def check_G_values(Gs, Gc, inp, max_index):
    if Gs > Gc:
        print(f"{inp[max_index]} is an outlier")
    else:
        print(f"{inp[max_index]} is not an outlier")
```

```python
Gcritical = calculate_critical_value(len(data['game_id']), 0.05)
Gstat, max_index = grubbs_stat(data['game_id'])
check_G_values(Gstat, Gcritical, data['game_id'], max_index)
```

**Тест Рознера**

Для теста Рознера достаточно дописать одну функцию, которая принимает набор данных, уровень значимости и число потенциальных выбросов

```python
def ESD_test(input_series, alpha, max_outliers):
    for iteration in range(max_outliers):
        Gcritical = calculate_critical_value(len(input_series), alpha)
        Gstat, max_index = grubbs_stat(input_series)
        check_G_values(Gstat, Gcritical, input_series, max_index)
        input_series = np.delete(input_series, max_index)
```

```python
# функция деления train test

def Xy(X):
    
    train = X.loc[~pd.isnull(X[target]),:]
    X_tr = train.drop([target], axis=1)
    y_tr = train[target]
         
    # целевая переменная является категориальной
    # переведем значения столбца в числа, оставив один столбец
    from sklearn.preprocessing import LabelEncoder
    le = LabelEncoder()
    # y = pd.Series(data = le.fit_transform(y))
    le.fit(y_tr)
    y_tr = pd.Series(data = le.transform(y_tr))
    
    from sklearn.model_selection import train_test_split
    X_train, X_test, y_train, y_test = train_test_split(X_tr, y_tr, test_size=0.3, random_state=0)
    
    return X_train, X_test, y_train, y_test
```

```python
X_train, X_test, y_train, y_test = Xy(X)
```

```python
y_test.shape
```

```python
# разница для каждого компонента
model = pca.fit(X_train)

def explained_variance(X):
    result = np.zeros(model.n_components)
    for ii in range(model.n_components):
        X_trans = model.transform(X)
        X_trans_ii = np.zeros_like(X_trans)
        X_trans_ii[:, ii] = X_trans[:, ii]
        X_approx_ii = model.inverse_transform(X_trans_ii)

        result[ii] = 1 - (np.linalg.norm(X_approx_ii - X) /
                          np.linalg.norm(X - model.mean_)) ** 2
    return result

print(model.explained_variance_ratio_[:5])
print(explained_variance(X_train)[:5])
print(explained_variance(X_test)[:5])
```

```python
#общая разъясненная разница
from sklearn.metrics import r2_score

print(model.explained_variance_ratio_.sum())
print(r2_score(X_train, model.inverse_transform(model.transform(X_train)),
               multioutput='variance_weighted'))
print(r2_score(X_test, model.inverse_transform(model.transform(X_test)),
               multioutput='variance_weighted'))
```

```python
# функции предсказания модели в классе

class p():
    
    def _init_(self, X_train, X_test, y_train, y_test):
        self= self

    def RandomForestClassifier(self, X_train, X_test, y_train, y_test):
        self.X_train = X_train
        self.X_test = X_test
        self.y_train = y_train
        self.y_test = y_test
        
        from sklearn.ensemble import RandomForestClassifier
        from sklearn.model_selection import GridSearchCV

        clf_rf = RandomForestClassifier(random_state=1)

        grid_params = {
                'criterion': ['gini'],     # функция измерения качества разделения
                'max_depth': [10],        # максимальная глубина дерева
                'n_estimators': [10],             # количество моделей (деревьев)
                'min_samples_leaf': [10], # мин число объектов в листе
                'max_features': [0.5]             # число признаков случайного подпространства
        }

        grid = GridSearchCV(clf_rf, param_grid=grid_params, cv=10, n_jobs=-1)
        # grid - исчерпывающий поиск по заданным значениям параметров
        # n_jobs - количество заданий для параллельного выполнения
        # cv - количество фолдов
        grid.fit(self.X_train, self.y_train)
        y_test_grid = grid.predict(self.X_test)
        self.y_test_grid = y_test_grid
        scor_grid = grid.score(self.X_test, self.y_test)
        self.scor_grid = scor_grid
        return grid.score(self.X_train, self.y_train), self.scor_grid, self.y_test_grid

    def LogisticRegression(self):

        from sklearn.linear_model import LogisticRegression
        from sklearn.preprocessing import StandardScaler
        from sklearn.pipeline import make_pipeline

        clf_log = make_pipeline(StandardScaler(),LogisticRegression(max_iter=1000, random_state=1))
        clf_log.fit(self.X_train, self.y_train)
        y_test_log = clf_log.predict(self.X_test)
        self.y_test_log = y_test_log
        scor_clf_log = clf_log.score(self.X_test, self.y_test)
        self.scor_clf_log = scor_clf_log

        return clf_log.score(self.X_train, self.y_train), self.scor_clf_log, self.y_test_log

    def L2_RidgeClassifier(self):  # для классификации  # RidgeCV
        
        from sklearn.linear_model import RidgeClassifier
        from sklearn.model_selection import cross_val_score
        from sklearn.preprocessing import StandardScaler
        from sklearn.pipeline import make_pipeline
        from statistics import mean

        cross_val_scores_ridge = []
        alpfa = []

        for i in range(1,10):
            clf_rg = make_pipeline(StandardScaler(),RidgeClassifier(alpha=i*0.25, normalize = False, random_state=1))
            clf_rg.fit(self.X_train, self.y_train)
            scores = cross_val_score(clf_rg, self.X_train, self.y_train, cv=10)
            avg_cross_val_score = mean(scores)*100
            cross_val_scores_ridge.append(avg_cross_val_score)
            alpfa.append(i*0.25)

        scor = dict(zip(alpfa, cross_val_scores_ridge)) 

        import operator
        alpfa_best = max(scor.items(), key=operator.itemgetter(1))[0] 

        clf_rg = make_pipeline(StandardScaler(),RidgeClassifier(alpha=alpfa_best))
        clf_rg.fit(self.X_train, self.y_train)

        y_test_rg = clf_rg.predict(self.X_test)
        self.y_test_rg = y_test_rg
        scor_rg = clf_rg.score(self.X_test, self.y_test)
        self.scor_rg = scor_rg
        return clf_rg.score(self.X_train, self.y_train), self.scor_rg,  self.y_test_rg 

    def L2(self):    # Ridge
        
        from sklearn.model_selection import GridSearchCV
        from sklearn.linear_model import LogisticRegression
        from sklearn.preprocessing import StandardScaler
        from sklearn.pipeline import make_pipeline
        
        self.X_train = StandardScaler().fit_transform(self.X_train)
        self.X_test = StandardScaler().fit_transform(self.X_test)
        
        clf_l2 = LogisticRegression(random_state=1)

        grid_params = {
                'penalty': ['l2'],        
                'C': np.arange(0.0, 1.0, 0.1),           
                'solver': ['sag', 'saga', 'liblinear']          
        }

        l2 = GridSearchCV(clf_l2, param_grid=grid_params, cv=10, n_jobs=-1)
        # grid - исчерпывающий поиск по заданным значениям параметров
        # n_jobs - количество заданий для параллельного выполнения
        # cv - количество фолдов
        l2.fit(self.X_train, self.y_train)
        y_test_l2 = l2.predict(self.X_test)
        self.y_test_l2 = y_test_l2
        scor_l2 = l2.score(self.X_test, self.y_test)
        self.scor_l2 = scor_l2
        return l2.score(self.X_train, self.y_train), self.scor_l2, self.y_test_l2
        
        
        
    def L1(self):   #Lasso

        from sklearn.model_selection import GridSearchCV
        from sklearn.linear_model import LogisticRegression
        from sklearn.preprocessing import StandardScaler   # X стандартизирован выше
        from sklearn.pipeline import make_pipeline
        
        self.X_train = StandardScaler().fit_transform(self.X_train)
        self.X_test = StandardScaler().fit_transform(self.X_test)
        
        clf_l1 = LogisticRegression(random_state=1)

        grid_params = {
                'penalty': ['l1'],        
                'C': np.arange(0.0, 1.0, 0.1),           
                'solver': ['sag', 'saga', 'liblinear']          
        }

        l1 = GridSearchCV(clf_l1, param_grid=grid_params, cv=10, n_jobs=-1)
        # grid - исчерпывающий поиск по заданным значениям параметров
        # n_jobs - количество заданий для параллельного выполнения
        # cv - количество фолдов
        l1.fit(self.X_train, self.y_train)
        y_test_l1 = l1.predict(self.X_test)
        self.y_test_l1 = y_test_l1
        scor_l1 = l1.score(self.X_test, self.y_test)
        self.scor_l1 = scor_l1
        return l1.score(self.X_train, self.y_train), self.scor_l1, self.y_test_l1
    
    #def SGDClassifier(self):    # код рабочий, но очень долго грузится
        
        #from sklearn.linear_model import SGDClassifier
        #from sklearn.model_selection import GridSearchCV
        #from sklearn.preprocessing import StandardScaler  # X стандартизирован выше
        
        #clf_ = SGDClassifier(random_state=1)
        
        #grid_params = {
            #'alpha': np.arange(0.001, 2.5, 0.01),
            #'loss': ['hinge', 'log', 'modified_huber'],
            #'penalty':['l2', 'l1', 'elasticnet'],
            #'l1_ratio':[0.15], 
            #'n_iter_no_change':[5],
            #'shuffle':[True], 
            #'average':[False], 
            #'class_weight':['balanced'],
            #'warm_start': [True],
            #'learning_rate': ['adaptive', 'optimal', 'invscaling'],
            #'eta0': [0.01, 0.5, 1]
        #}
        
        #clf_sgd = GridSearchCV(clf_, param_grid=grid_params, cv=10, n_jobs=-1, 
                               #scoring='f1_macro', verbose=0, refit=True)

        #clf_sgd.fit(self.X_train, self.y_train)
        #y_test_sgd = clf_sgd.predict(self.X_test)
        #self.y_test_sgd = y_test_sgd
        #scor_sgd = clf_sgd.score(self.X_test, self.y_test)
        #self.scor_sgd = scor_sgd
        #return clf_sgd.score(self.X_train, self.y_train), self.scor_sgd, self.y_test_sgd     
               
    def table(self):
        mapping = {'RandomForestClassifier':self.scor_grid, 'LogisticRegression':self.scor_clf_log, 
                   'RidgeClassifier':self.scor_rg, 'L2':self.scor_l2, 'L1': self.scor_l1}
        for i, j in mapping.items():
            print(f'{i}: {j}')
    
    def plot_model(self):
        models = ['RandomForestClassifier', 'LogisticRegression', 'RidgeClassifier', 'L2', 'L1']
        scores = [self.scor_grid, self.scor_clf_log, self.scor_rg, self.scor_l2, self.scor_l1]
        plt.bar(models, scores)
        plt.xlabel('Model')
        plt.ylabel('Score')
        plt.grid
        plt.show()
```
